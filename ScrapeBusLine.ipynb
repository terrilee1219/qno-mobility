```python
import requests
from bs4 import BeautifulSoup
import urllib.request

url ="https://www.myciti.org.za/en/routes-stops/route-stop-station-map/"

response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

'''print(soup.prettify())'''

# divLine is an array
divLine = soup.find_all("div", attrs={"class":"result routes-stops-map-route"})

# open a text file
f = open("BusLine.txt", "w")

#print(divLine)
for element in divLine:
    spanLine = element.find_all("span")  # spanLine is an array of all the span tags in each element
    resultTag = spanLine[0].contents[0]
    resultTitle = spanLine[1].contents[0]
    # put above into a single string
    result = str(resultTag)+" "+str(resultTitle)+"\n"
    print(result)
    f.write(result)
f.close()
'''
for element in range(2):
    print(spanLine[element].contents[0])
'''
```
